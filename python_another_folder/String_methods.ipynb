```python
| Function       | Example                                        | Explanation                           |
| -------------- | ---------------------------------------------- | ------------------------------------- |
| `lower()`      | `"HeLLo".lower()` → `'hello'`                  | Converts all characters to lowercase  |
| `upper()`      | `"hello".upper()` → `'HELLO'`                  | Converts all characters to uppercase  |
| `title()`      | `"hello world".title()` → `'Hello World'`      | Capitalizes first letter of each word |
| `capitalize()` | `"hello world".capitalize()` → `'Hello world'` | Capitalizes first character only      |
| `swapcase()`   | `"HeLLo".swapcase()` → `'hEllO'`               | Swaps uppercase ↔ lowercase           |
| `casefold()`   | `"HELLO".casefold()` → `'hello'`               | Aggressive lowercasing (Unicode-safe) |
```

```python
| Function   | Example                     | Explanation                                                       |
| ---------- | --------------------------- | ----------------------------------------------------------------- |
| `find()`   | `"hello".find("l")` → `2`   | Returns index of substring (first match). Returns -1 if not found |
| `rfind()`  | `"hello".rfind("l")` → `3`  | Search from the right side                                        |
| `index()`  | `"hello".index("e")` → `1`  | Same as find(), but throws error if not found                     |
| `rindex()` | `"hello".rindex("l")` → `3` | Same as rfind(), but throws error                                 |
| `count()`  | `"banana".count("a")` → `3` | Counts how many times substring appears                           |
```

```python
| Function         | Example                            | Explanation                                 |
| ---------------- | ---------------------------------- | ------------------------------------------- |
| `isalpha()`      | `"abc".isalpha()` → `True`         | Only alphabets                              |
| `isdigit()`      | `"123".isdigit()` → `True`         | Only digits                                 |
| `isalnum()`      | `"abc123".isalnum()` → `True`      | Alphabets or digits                         |
| `islower()`      | `"hello".islower()` → `True`       | All lowercase                               |
| `isupper()`      | `"HELLO".isupper()` → `True`       | All uppercase                               |
| `isspace()`      | `"   ".isspace()` → `True`         | Only whitespace                             |
| `istitle()`      | `"Hello World".istitle()` → `True` | Title format                                |
| `isdecimal()`    | `"123".isdecimal()` → `True`       | Decimal numbers only                        |
| `isdigit()`      | `"123".isdigit()` → `True`         | Digits (includes unicode)                   |
| `isnumeric()`    | `"Ⅻ".isnumeric()` → `True`         | Numeric characters including roman numerals |
| `isidentifier()` | `"name_1".isidentifier()` → `True` | Valid Python variable name                  |
| `isascii()`      | `"A".isascii()` → `True`           | ASCII characters only                       |
```

```python
| Function   | Example                     | Explanation                              |
| ---------- | --------------------------- | ---------------------------------------- |
| `strip()`  | `"  hi  ".strip()` → `'hi'` | Removes spaces (or chars) from both ends |
| `lstrip()` | `"  hi".lstrip()` → `'hi'`  | Removes left side spaces                 |
| `rstrip()` | `"hi  ".rstrip()` → `'hi'`  | Removes right side spaces                |
```

```python
| Function       | Example                                                       | Explanation           |
| -------------- | ------------------------------------------------------------- | --------------------- |
| `replace()`    | `"hello world".replace("world", "python")` → `'hello python'` | Replace substring     |
| `split()`      | `"a,b,c".split(",")` → `['a','b','c']`                        | Split into list       |
| `rsplit()`     | `"a,b,c".rsplit(",",1)` → `['a','b','c']`                     | Split from right      |
| `splitlines()` | `"a\nb".splitlines()` → `['a','b']`                           | Split by newline      |
| `join()`       | `" - ".join(['a','b'])` → `'a - b'`                           | Join list into string |
```

```python
| Function       | Example                                | Explanation                |
| -------------- | -------------------------------------- | -------------------------- |
| `startswith()` | `"hello".startswith("he")` → `True`    | Checks beginning of string |
| `endswith()`   | `"file.txt".endswith(".txt")` → `True` | Checks end of string       |
```

```python
| Function       | Example                                         | Explanation                 |
| -------------- | ----------------------------------------------- | --------------------------- |
| `format()`     | `"{} is {}".format("Python", "easy")`           | Format replacement fields   |
| `format_map()` | `"Name: {name}".format_map({'name': 'Sachin'})` | Formatting using dict       |
| `f-string`     | `f"Hi {name}"`                                  | Fastest & modern formatting |
```

```python
| Function   | Example                     | Explanation       |
| ---------- | --------------------------- | ----------------- |
| `center()` | `"hi".center(5)` → `' hi '` | Center align text |
| `ljust()`  | `"hi".ljust(5)` → `'hi   '` | Left align        |
| `rjust()`  | `"hi".rjust(5)` → `'   hi'` | Right align       |
| `zfill()`  | `"42".zfill(5)` → `'00042'` | Pads with zeros   |
```

```python
| Function       | Example                                                       | Explanation        |
| -------------- | ------------------------------------------------------------- | ------------------ |
| `partition()`  | `"hello-world".partition("-")` → `('hello','-','world')`      | Split into 3 parts |
| `rpartition()` | `"hello-abc-xyz".rpartition("-")` → `('hello-abc','-','xyz')` | Split from right   |
```
