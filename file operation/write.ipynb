```python
# ✔️ What it does

# write() writes only a string to a file.
# If the file already exists, its content will be overwritten (when opened in "w" mode).

# Syntax:
        # file.write(string)
```

```python
with open("example.txt", "w") as f:
    f.write("Hello World\n")
    f.write("Welcome to Python file handling.")
```

```python
# ✔️ Important points
# You must manually add newline character \n if you want a new line.
# You can call write() multiple times.
# It returns the number of characters written.

with open("a.txt", "w") as f:
    n = f.write("Hi")
    print(n)  # Output: 2
```

```python
# ✍️ 2. The writelines() Method
# ✔️ What it does
# Writes multiple strings (e.g., list of strings) into the file.

# ✔️ Example
lines = ["Arun\n", "Megha\n", "Ravi\n"]

with open("names.txt", "w") as f:
    f.writelines(lines)
```

```python
# ✍️ 3. File Modes for Writing
# (a) "w" → Write Mode
# Creates new file if not exists
# Overwrites content if file exists
# Example:
with open("data.txt", "w") as f:
    f.write("New content")
```

```python
# (b) "a" → Append Mode

# Does NOT remove old content

# Adds new text at the end of file

# Example:

with open("data.txt", "a") as f:
    f.write("\nThis line is appended.")
```

```python
# (c) "w+" → Write + Read
# Overwrites file
# Allows reading also
# Example:

with open("data.txt", "w+") as f:
    f.write("Hello\n")
    f.seek(0)
    print(f.read())
```

```python
# (d) "a+" → Append + Read
# Keeps old data
# Adds new data
# You can read after seek(0)
# Example:

with open("data.txt", "a+") as f:
    f.write("\nNew entry")
    f.seek(0)
    print(f.read())
```

```python
# ✍️ 4. Creating a File if Not Exists

# Any write mode "w", "a", "w+", "a+"
# → automatically creates a new file if it doesn't exist.
# ✍️ 5. Writing Non-String Data (Important)

# You must convert objects to string.

# Example:

n = 100
with open("number.txt", "w") as f:
    f.write(str(n))
```
