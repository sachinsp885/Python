```python
#simple read function using 'r' inside open function we can read data from file.
# f.read() ->whole content of the data
# f.read(10) -> its we are reading the first ten lines from the file
# f.readline() ->it will read first line at a time
# f.readlines() ->it will create the list
```

```python
# ‚úÖ 1. read()
# Reads the entire file at once (as one single string).
# Example:
with open("sample.txt", "r") as f:
    data = f.read()
    print(data)

# ‚úîÔ∏è Good when:
# File is small
# You need the full content at once
# ‚ùå Not good when:
# File is large (memory issue)    
```

```python
# ‚úÖ 2. read(n)
# Reads first n characters from the file.
# Useful to read file in chunks.
with open("sample.txt", "r") as f:
    data = f.read(10)   # reads first 10 characters
    print(data)
```

```python
# ‚úÖ 3. readline()
# Reads one line at a time.
# Example:
with open("sample.txt", "r") as f:
    line1 = f.readline()
    print(line1)

    line2 = f.readline()
    print(line2)

# ‚úîÔ∏è Useful when:
# You want line-by-line processing
# Memory efficient
```

```python
# ‚úÖ 4. readlines()
# Reads all lines and returns a list of strings.
# Example:
with open("sample.txt", "r") as f:
    lines = f.readlines()
    print(lines)

# ‚úîÔ∏è Good for:
# When you want to loop easily through all lines
# ‚ùå Downside:
# Loads entire file into memory
```

```python
# ‚úÖ 5. Looping through file directly
# Most common and memory-efficient.
# Example
with open("sample.txt", "r") as f:
    for line in f:
        print(line.strip())   # strip removes newline

# ‚úîÔ∏è Very efficient
# No need to store entire file in memory.
```

```python
# ‚úÖ 6. Using .seek() and .tell() (advanced but useful)
# tell() ‚Üí Shows current cursor position
# seek() ‚Üí Moves cursor to a specific position

with open("sample.txt", "r") as f:
    print(f.tell())   # 0 at beginning

    data = f.read(5)
    print(data)

    print(f.tell())   # now 5

    f.seek(0)         # move back to start
    print(f.read())
```

```python
# üéØ Summary Table
| Method                    | Description         | Memory Usage |
| ------------------------- | ------------------- | ------------ |
| `read()`                  | Full file as string | High         |
| `read(n)`                 | First n characters  | Low          |
| `readline()`              | One line            | Very low     |
| `readlines()`             | All lines as list   | High         |
| Loop (`for line in file`) | Line-by-line        | Very low     |
| `seek() / tell()`         | Move cursor         | N/A          |
------------------------------------------------------------------
```

```python
with open('sample.txt','r') as f:
    print(f.read())

with open('sample.txt','r') as f:
    print(f.readline())

with open('sample.txt','r') as f:
    print(f.readlines())
```

```python
with open('sample.txt', 'r') as f:
    for line in f:
        print(line.strip())
```
