```python
import pandas as pd

# Sample DataFrame
data = {
    'EmpID': [101,102,103,104,105,106,107,108,109,110],
    'Name': ['Arun','Meena','Kiran','Ravi','Anita','Sanjay','Pooja','Rohit','Divya','Sachin'],
    'Department': ['IT','HR','Finance','IT','HR','Finance','IT','Finance','HR','IT'],
    'Project': ['Alpha','Bravo','Alpha','Delta','Echo','Delta','Alpha','Echo','Bravo','Delta'],
    'HoursWorked': [120,90,160,140,110,170,150,130,95,165],
    'Experience': [3,5,2,6,4,7,2,5,3,8],
    'Rating': [4.5,4.2,4.0,4.3,4.1,4.7,3.9,4.4,4.0,4.8],
    'Salary': [50000,60000,55000,70000,45000,75000,52000,68000,47000,72000],
    'city':['Mumbai','Delhi','Mumbai','Delhi','Mumbai','Delhi','Mumbai','Delhi','Mumbai','Delhi']
}

df = pd.DataFrame(data)

# Display initial DataFrame
# print("Initial DataFrame:")
# print(df)

try:
    df["Bonus"] = df["Salary"].apply(lambda x:x+(0.1*x) if x>60000 else (0.05*x)+x )
except Exception as e:
    print("error in 1 question")

# 2. Add a new column 'TaxedSalary' = Salary - Tax
#    - Tax = 10% of Salary if Salary > 50000
#    - Tax = 5% of Salary otherwise
try:
    df["TaxedSalary"] = df["Salary"].apply(lambda x:x-(0.1*x) if x>50000 else x-(0.05*x))
except Exception as e:
    print("error in 2 question")

# 3. Add a column 'Seniority' based on Experience:
#    - Experience >= 5 → 'Senior'
#    - Experience 3–4 → 'Mid'
#    - Experience < 3 → 'Junior'
try:
    df['Seniority'] = df['Experience'].apply(lambda x: 
                                             'Senior' if x>=5 else 
                                             'Mid' if x>=3 else 'Junior')
except Exception as e:
    print("error in 3 question")

# 4. Add a column 'CityGroup' based on City:
#    - If City in ['Mumbai','Delhi'] → 'Metro'
#    - Else → 'Other'

try:
    df['CityGroup'] = df['City'].apply(lambda x:'Metro' if x in ['Mumbai','Delhi'] else 'other')
except Exception as e:
    print(f"error in 4 question")

# 5. Add a column 'PerformanceScore' = Rating * Experience
try:
    df['PerformanceScore'] = df['Rating'] * df['Experience']
except Exception as e:
    print("error in 5 question")

# 6. Add a column 'ExtraPay' based on HoursWorked:
#    - HoursWorked > 150 → 2000
#    - Else → 1000
try:
    df['ExtraPay'] = df['HoursWorked'].apply(lambda x:2000 if x>150 else 1000)
except Exception as e:
    print("error in 6 question")

# 7. Add a column 'AdjustedSalary' = Salary + Bonus + ExtraPay
try:
    df['AdjustedSalary'] = df['Salary'] + df['Bonus'] + df["ExtraPay"]
except Exception as e:
    print("error in 7 question")

# 8. Add a column 'HighEarner':
#    - True if AdjustedSalary > 70000
#    - Else False
try:
    df['HighEarner'] = df['AdjustedSalary'].apply(lambda x:'True' if x>70000 else 'False')
except Exception as e:
    print("error in 8 question")

# 9. Add a column 'ProjectLoad':
#    - If HoursWorked > 140 and Project in ['Alpha','Delta'] → 'Heavy'
#    - Else → 'Normal'

try:
    df['ProjectLoad'] = df.apply(lambda x: 'Heavy' 
                                 if x['HoursWorked']>140 and x['Project'] in ['Alpha','Delta'] else 'Normal', axis=1)

except Exception as e:
    print("error in 9 question")


# 10. Add a column 'RatingLevel' based on Rating:
#    - Rating >= 4.5 → 'Excellent'
#    - Rating 4–4.4 → 'Good'
#    - Rating < 4 → 'Average'
try:
        df['RatingLevel'] = df['Rating'].apply(lambda x: 'Excellent' 
                                               if x>=4.5 else 'Good' 
                                               if x>=4 else 'Average')

except Exception as e:
    print("error in 10 question")
```
