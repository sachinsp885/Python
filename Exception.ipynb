```python
# Exception handling is used to handle errors in a program without stopping the code.
# Python uses try–except blocks.
# ✅ Basic Syntax
try:
    a=10/0 #risk code
except ZeroDivisionError:
    print('zero')   # handle error
```

```python
# ✅ Handling Specific Exceptions
try:
    a = int("abc")
except ValueError:
    print("Invalid number")
```

```python
# ✅ Multiple Except Blocks
try:
    x = 10 / int(input("Enter number: "))
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Enter a valid number")
else:
    print(x)
finally:
    print("End I am here")
```

```python
# ✅ Raise Your Own Exception
age = -2

if age < 0:
    raise Exception("Age cannot be negative")
```

```python
# ✅ Using except Exception as e

# This captures the actual error message into variable e.

# ✔ Example
try:
    x = 10 / 0
except Exception as e:
    print("Error:", e)
```
