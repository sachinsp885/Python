```python
from collections import Counter

d1 = {"a": 1, "b": 2}
d2 = {"b": 3, "d": 4}

merged = dict(Counter(d1) + Counter(d2))
print(merged)
```

```python
d1 = {"a": 1, "b": 2}
d2 = {"b": 3, "d": 4}

merged = d1.copy()

for k, v in d2.items():
    if k in merged:
        merged[k] += v
    else:
        merged[k] = v

print(merged)
```

```python
students = {
    "Sachin": 85,
    "Rohit": 92,
    "Anu": 85,
    "Manu": 78,
    "Kiran": 92,
    "Ravi": 60
}

#1st solution
result ={}
for name,marks in students.items():
    result.setdefault(marks,[]).append(name)
print(result)

#2nd solution
from collections import defaultdict
result = defaultdict(list)
for name, marks in students.items():
    result[marks].append(name)
result = dict(result)

#3rd solution
result = {}
for name, marks in students.items():
    if marks not in result:
        result[marks] = []
    result[marks].append(name)
```

```python
data = {
    "Sachin": {"math": 90, "science": 85},
    "Rohit": {"math": 78, "science": 92},
    "Kiran": {"math": 88, "science": 80}
}
output ={}
for name,sub_marks in data.items():
    if isinstance(sub_marks,dict):
        for sub,marks in sub_marks.items():
            output.setdefault(sub,[]).append(marks)
print(output)  


from collections import defaultdict
output = defaultdict(list)
for subjects in data.values():
    for sub, marks in subjects.items():
        output[sub].append(marks)
output = dict(output)

output = {}
for subjects in data.values():
    for sub, marks in subjects.items():
        if sub not in output:
            output[sub] = []
        output[sub].append(marks)
```

```python
records = {
    "t1": {"city": "Bangalore", "temp": 28},
    "t2": {"city": "Mumbai", "temp": 32},
    "t3": {"city": "Bangalore", "temp": 30},
    "t4": {"city": "Delhi", "temp": 25}
}

temp_sum = {}
temp_count = {}
output = {}

for rec, details in records.items():
    city = details["city"]
    temp = details["temp"]

    temp_sum[city] = temp_sum.get(city, 0) + temp
    temp_count[city] = temp_count.get(city, 0) + 1

for city in temp_sum:
    output[city] = temp_sum[city] // temp_count[city]   # average

print(output)
```

```python
employees = [
    {"name": "Sachin", "dept": "IT", "salary": 70000},
    {"name": "Rohit", "dept": "HR", "salary": 50000},
    {"name": "Kiran", "dept": "IT", "salary": 80000},
    {"name": "Manu", "dept": "Finance", "salary": 60000},
    {"name": "Anu", "dept": "HR", "salary": 55000},
]

output = {}
for emp in employees:
    dept = emp["dept"]
    salary = emp["salary"]
    if dept not in output:
        output[dept] = salary
        print(output)
    else:
        output[dept] = max(output[dept], salary)

print(output)
```

```python
sales = [
    {"product": "Laptop", "qty": 2, "price": 50000},
    {"product": "Mouse", "qty": 10, "price": 500},
    {"product": "Laptop", "qty": 1, "price": 50000},
    {"product": "Keyboard", "qty": 3, "price": 1200},
    {"product": "Mouse", "qty": 5, "price": 500}
]

# {
#     "Laptop": 150000,
#     "Mouse": 7500,
#     "Keyboard": 3600
# }

output = {}
for sale in sales:
    products = sale["product"]
    prices = sale["price"]
    qunties = sale['qty']
    if products not in output:
        output[products] = prices * qunties
    else:
        output[products] += prices * qunties

print(output)

```
