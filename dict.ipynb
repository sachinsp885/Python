```python
marks = {"math": 80, "science": 92, "english": 75}
max_marks = max(marks.values())
for sub,sub_marks in marks.items():
    if  sub_marks == max_marks:
        print(sub)
```

```python
d1 = {"a": 1, "b": 2}
d2 = {"c": 3, "d": 4}
output = d1.copy()
for i,j in d2.items():
    output.update({i:j})
print(output)
```

```python
d = {"apple": 5, "banana": 2, "cherry": 8}

# sorted returns a list of tuples
sorted_d = dict(sorted(d.items(), key=lambda item: item[1]))

print(sorted_d)
```

```python
from collections import Counter

d1 = {"a": 1, "b": 2}
d2 = {"b": 3, "d": 4}

merged = dict(Counter(d1) + Counter(d2))
print(merged)
```

```python
d1 = {"a": 1, "b": 2}
d2 = {"b": 3, "d": 4}

merged = d1.copy()

for k, v in d2.items():
    if k in merged:
        merged[k] += v
    else:
        merged[k] = v

print(merged)
```
